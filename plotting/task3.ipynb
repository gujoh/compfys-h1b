```python
# Imports
import numpy as np 
import matplotlib.pyplot as plt
import scipy
import warnings
warnings.filterwarnings("ignore")
```

```python
data0001 = np.genfromtxt("../data/task3_1.0e-03.csv", delimiter=",")
datas = [data0001]
dts = [0.001]
```

```python
def plot(data, t, dt):
    fig, ax = plt.subplots(ncols=2, figsize=(13, 6))
    fig.suptitle(f"Energy and temperature plotted against time, dt = {dt}")
    ax[0].plot(t, data[:, 0], label="Potential")
    ax[0].plot(t, data[:, 1], label="Kinetic")
    ax[0].plot(t, data[:, 0] + data[:, 1], label="Total")
    ax[0].set_xlabel("Time (ps)")
    ax[0].set_ylabel("Energy (eV)")
    ax[0].legend()
    ax[1].plot(t, np.cumsum(data[:, 2]) / (t / dt), label="Temperature", c="purple")
    ax[1].set_xlabel("Time (ps)")
    ax[1].set_ylabel("Temperature (K)")
    ax[1].legend()
    plt.show()
```

```python
for data, dt in zip(datas, dts):
    t = np.arange(0, data.shape[0]) * dt
    print(f"Average temperature: {np.mean(data[:, 2]):.4f}K")
    print(f"Average pressure: {np.mean(data[:, 3]):.4f}")
    plot(data, t, dt)

plt.plot(t, data0001[:, 3])
plt.show()
```
