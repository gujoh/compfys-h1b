```python
# Imports
import numpy as np 
import matplotlib.pyplot as plt
import scipy
import warnings
warnings.filterwarnings("ignore")
```

```python
data = np.genfromtxt("../data/task2.csv", delimiter=",")
dt = 0.001
t = np.arange(0, data.shape[0]) * dt
```

```python
plt.plot(t, data[:, 0], label="Potential")
plt.plot(t, data[:, 1], label="Kinetic")
plt.plot(t, np.sum(data, axis=1), label="Total")
plt.xlabel("Time (ps)")
plt.legend()
plt.show()
```
