```python
# Imports
import numpy as np 
import matplotlib.pyplot as plt
import scipy
import warnings
warnings.filterwarnings("ignore")
```

```python
data01 = np.genfromtxt("../data/task2_1.0e-01.csv", delimiter=",")
data2001 = np.genfromtxt("../data/task2_2.0e-02.csv", delimiter=",")
#data3001 = np.genfromtxt("../data/task2_3.0e-02.csv", delimiter=",")
data0001 = np.genfromtxt("../data/task2_1.0e-03.csv", delimiter=",")
datas = [data01, data2001, data0001]
dts = [0.1, 0.02, 0.001]
```

```python
def plot(data, t, dt):
    fig, ax = plt.subplots(ncols=2, figsize=(13, 6))
    #fig.suptitle(f"Energy and temperature plotted against time, dt = {dt}")
    ax[0].plot(t, data[:, 0], label=f"Potential, $\Delta$t={dt}ps")
    ax[0].plot(t, data[:, 1], label=f"Kinetic, $\Delta$t={dt}ps")
    ax[0].plot(t, data[:, 0] + data[:, 1], label=f"Total, $\Delta$t={dt}ps")
    ax[0].set_xlabel("Time (ps)")
    ax[0].set_ylabel("Energy (eV)")
    ax[0].legend()
    #ax[1].plot(t, np.cumsum(data[:, 2]) / (t / dt), label="Temperature", c="purple")
    ax[1].plot(t, data[:, 2], label=f"Temperature, $\Delta$t={dt}ps")
    print(np.mean(data[2000:,2]))
    ax[1].set_xlabel("Time (ps)")
    ax[1].set_ylabel("Temperature (K)")
    ax[1].legend()
    plt.show()
```

```python
for data, dt in zip(datas, dts):
    t = np.arange(0, data.shape[0]) * dt
    plot(data, t, dt)
```

```python
color_iterator = iter(['#8931EF', '#F2CA19', '#0057E9', '#FF00BD', '#87E911', '#E11845'])
fig, ax = plt.subplots(ncols=2, figsize=(13, 6))
datas2 = [data01, data2001]
dts2 = [0.1, 0.02]
for data, dt in zip(datas2, dts2):
    t = np.arange(0, data.shape[0]) * dt
    ax[0].plot(t, data[:, 0], label=f"Potential, $\Delta$t={dt} ps", color=next(color_iterator))
    ax[0].plot(t, data[:, 1], label=f"Kinetic, $\Delta$t={dt} ps", linewidth = 5, color=next(color_iterator))
    ax[0].plot(t, data[:, 0] + data[:, 1], label=f"Total, $\Delta$t={dt} ps", color=next(color_iterator))
    ax[0].set_xlabel("Time (ps)")
    ax[0].set_ylabel("Energy (eV)")
    ax[1].plot(t, data[:, 2], label=f"Temperature, $\Delta$t={dt} ps")
    ax[1].set_xlabel("Time (ps)")
    ax[1].set_ylabel(f"Temperature (K)")
ax[0].legend()
ax[1].legend()
plt.show()
```
