```python
# Imports
import numpy as np 
import matplotlib.pyplot as plt
import scipy
import warnings
warnings.filterwarnings("ignore")
```

```python
data01 = np.genfromtxt("../data/task2_1.0e-01.csv", delimiter=",")
data001 = np.genfromtxt("../data/task2_1.0e-02.csv", delimiter=",")
data3001 = np.genfromtxt("../data/task2_3.0e-02.csv", delimiter=",")
data0001 = np.genfromtxt("../data/task2_1.0e-03.csv", delimiter=",")
datas = [data01, data3001, data001, data0001]
dts = [0.1, 0.03, 0.01, 0.001]
```

```python
def plot(data, t, dt):
    fig, ax = plt.subplots(ncols=2, figsize=(13, 6))
    fig.suptitle(f"Energy and temperature plotted against time, dt = {dt}")
    ax[0].plot(t, data[:, 0], label="Potential")
    ax[0].plot(t, data[:, 1], label="Kinetic")
    ax[0].plot(t, data[:, 0] + data[:, 1], label="Total")
    ax[0].set_xlabel("Time (ps)")
    ax[0].set_ylabel("Energy (eV)")
    ax[0].legend()
    ax[1].plot(t, np.cumsum(data[:, 2]) / t, label="Temperature", c="purple")
    ax[1].set_xlabel("Time (ps)")
    ax[1].set_ylabel("Temperature (eV)")
    ax[1].legend()
    plt.show()
```

```python
for data, dt in zip(datas, dts):
    t = np.arange(0, data.shape[0]) * dt
    plot(data, t, dt)
```
