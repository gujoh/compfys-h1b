```python
# Imports
import numpy as np 
import matplotlib.pyplot as plt
from scipy.optimize import minimize 
import warnings
warnings.filterwarnings("ignore")
```

```python
data = np.genfromtxt("../data/task1.csv", delimiter=",")
```

```python
plt.plot(data[:, 0], data[:, 1], "-.")
plt.xlabel("Unit cell volume $Å^3$")
plt.ylabel("Energy (eV/unit cell)")
plt.show()
```

Doing a quadratic fit

```python
fit = np.polyfit(data[:, 0], data[:, 1], deg=2)
y = lambda x :fit[0] * x ** 2 + fit[1] * x + fit[2]

min_val = minimize(y, x0=65)
min_x, min_y = min_val.x, y(min_val.x)
print(np.round(min_x[0]**(1/3),5))
min_label = f'$\min(quadratic\;fit) = ({np.round(min_val.x[0],3)},{np.round(y(min_val.x[0]),3)})$'

indicies = [x for x in range(0,1000,30)] # Plotting every datapoint is messy.

plt.plot(data[:,0], y(data[:,0]), '-',label='Quadratic fit',color='orange',linewidth=2)
plt.plot(data[indicies, 0], data[indicies, 1], "x", label= 'Data points', color='black',markersize=7)
plt.plot(min_x, min_y,'o', markersize=10, label=min_label)

plt.xlabel("Unit cell volume $Å^3$")
plt.ylabel("Energy (eV/unit cell)")

plt.legend()
plt.show()
```
