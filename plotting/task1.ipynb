```python
# Imports
import numpy as np 
import matplotlib.pyplot as plt
import scipy
import warnings
warnings.filterwarnings("ignore")
```

```python
data = np.genfromtxt("../data/task1.csv", delimiter=",")
```

```python
plt.plot(data[:, 0], data[:, 1], "-.")
plt.xlabel("Unit cell volume $Ã…^3$")
plt.ylabel("Energy (eV/unit cell)")
plt.show()
```

Doing a quadratic fit

```python
fit = np.polyfit(data[:, 0], data[:, 1], deg=2)
y = lambda x :fit[0] * x ** 2 + fit[1] * x + fit[2]
solved = scipy.optimize.fsolve(y, [64, 66])
print(f"Solution 1: {solved[0]:.5f}, solution 2:  {solved[1]:.5f}")
print(f"Lattice constants at K = 0: {solved[0] ** (1 / 3):.5f}, and {solved[1] ** (1 / 3):.5f}")
```
